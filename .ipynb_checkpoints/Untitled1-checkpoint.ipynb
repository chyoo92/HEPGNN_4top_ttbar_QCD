```python
### id abs

st6_r = []
st1_r = []
st_num1 = []
for i in range(len(f['events/px'])):
    px_6 = f['events/px'][i][np.abs(f['events/id'][i]) == 6]
    py_6 = f['events/py'][i][np.abs(f['events/id'][i]) == 6]
    
    px_1 = f['events/px'][i][np.abs(f['events/id'][i]) == 1]
    py_1 = f['events/py'][i][np.abs(f['events/id'][i]) == 1]
    
    r_6 = np.sqrt(px_6**2 + py_6**2).sum()
    r_1 = np.sqrt(px_1**2 + py_1**2).sum()
    
    
    st6_r.append(r_6)
    st1_r.append(r_1)
    st_num1.append(len(py_1))
```

```python

    
plt.hist(np.array(st6_r)[f_pred['weight']==1], 
     density=True, bins=100, log=True,
     color='blue',label='no weight',histtype = 'step')
plt.hist(np.array(st6_r)[f_pred['weight']==-1], 
     density=True, bins=100, log=True,
     color='red',label='no weight',histtype = 'step')
plt.legend(loc='best',fontsize=20)
plt.show()    
plt.hist(np.array(st1_r)[f_pred['weight']==1], 
     density=True, bins=100, log=True,
     color='blue',label='no weight',histtype = 'step')
plt.hist(np.array(st1_r)[f_pred['weight']==-1], 
     density=True, bins=100, log=True,
     color='red',label='no weight',histtype = 'step')
plt.legend(loc='best',fontsize=20)
plt.show() 
plt.hist(np.array(st_num1)[f_pred['weight']==1], 
     density=True, bins=6, log=True,
     color='blue',label='no weight',histtype = 'step')
plt.hist(np.array(st_num1)[f_pred['weight']==-1], 
     density=True, bins=6, log=True,
     color='red',label='no weight',histtype = 'step')
plt.legend(loc='best',fontsize=20)
plt.show()
```
